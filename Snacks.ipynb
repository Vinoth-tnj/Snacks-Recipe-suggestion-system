```python
df1=pd.get_dummies(snacks.name)
df1.head(2)
```

```python
X = df1
X.head(3)
```

```python
y = snacks.ingredients
y.head(3)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder= LabelEncoder()
y=encoder.fit_transform(y)
y_list = dict(zip(list(y), snacks['ingredients'].to_list()))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.tree import DecisionTreeRegressor
lr_clf = DecisionTreeRegressor()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

```python
def predict_price(names):
    loc_index = np.where(X.columns==names)[0][0]

    x = np.zeros(len(X.columns))
    if loc_index >= 0:
        x[loc_index] = 1

    return lr_clf.predict([x])[0]
```

```python
print(y_list[predict_price('Samosa')])
```
